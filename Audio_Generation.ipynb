This is a basic program on how we can generate different audio signals considering the parameters for creating a sound signal as a waveform.

```
# import modules
import numpy as np
from IPython.display import Audio
from scipy.io import wavfile
```

```
# Set to True if we are in colaboratory or False elsewhere
colaboratory = True

if colaboratory:
    # Load the calc module from the given link
    !rm calc.*
    !wget https://raw.githubusercontent.com/R6500/Python-bits/master/Modules/calc.py

# Now, import the module
import calc

# Clear output after import
from IPython.display import clear_output
clear_output()

# Check loaded module
try:
    print('calc version: ',calc.version)
except:
    print('Error loading the calc module')
```

```
# Generate a sound
framerate = 45000
t = np.linspace(0,8,framerate*5)
data = np.sin(2*np.pi*220*t) + np.sin(2*np.pi*225*t)
```

```
# Indicate the calc module if we are inside colaboratory
calc.setColaboratory(colaboratory)

# Show waveform
calc.plot11(t,data,"Sound Waveform","Time in sec","Value")

# We can zoom the start of the waveform
calc.plot11(t[0:10000],data[0:10000],"Sound Waveform (Detail)","Time in sec","Value")
```

```
# Generate a player for mono sound
Audio(data,rate=framerate,autoplay=True)
```

Now, you can manipulate the waveform to get different sound waves and thus, depending on that different audio will be generated.

```
# Take a 1s piece of the mono sound
data2 = data[0:framerate]

# Generate a player for mono sound
Audio(data2,rate=framerate,autoplay=True)
```
